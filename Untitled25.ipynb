```python
import re
from tqdm import tqdm
from typing import Any, List, Dict

class CustomSQLDataset(DatasetSpecificProcessing):

    def dataset_to_jsonl(self, dataset_jsonl: str, **kwargs: Any) -> None:
        def extract_sql_query(completion: str) -> str:
            """
            Extracts the SQL query from the answer string.
            This version assumes the answer field is mostly SQL with minimal noise.
            """
            if not completion:
                return self.INVALID_ANS
            return completion.strip()

        examples_set: List[Dict[str, Any]] = []

        for _, sample in tqdm(enumerate(kwargs["dataset"]), desc="Processing SQL QA Samples"):
            question = sample.get("question", "").strip()
            schema = sample.get("table_schema", "").strip()
            answer = sample.get("answer", "").strip()

            example = {
                DatasetSpecificProcessing.QUESTION_LITERAL: question,
                "table_schema": schema,
                DatasetSpecificProcessing.ANSWER_WITH_REASON_LITERAL: answer,
                DatasetSpecificProcessing.FINAL_ANSWER_LITERAL: extract_sql_query(answer)
            }
            examples_set.append(example)

        save_jsonlist(dataset_jsonl, examples_set, "w")

    def extract_final_answer(self, answer: str) -> str:
        """
        Optional cleaner if answers include commentary + SQL
        """
        if not answer:
            return self.INVALID_ANS

        # Extract the first SQL-like statement using naive detection
        sql_match = re.search(r"(SELECT .*?)(;|\n|$)", answer, re.IGNORECASE | re.DOTALL)
        if sql_match:
            return sql_match.group(1).strip()

        return self.INVALID_ANS
```
